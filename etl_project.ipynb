```python
# Set up dependencies
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
# Reference the file where the CSV is located
top50_path = 'top50_2019.csv'
# Import the data into a Pandas DataFrame
top50_df = pd.read_csv(top50_path)
top50_df.head()
```

```python
# update df to keep specific columns and sort by speechiness
cleaned_top50 = top50_df[['Track_Name', 'Artist_Name', 'Genre', 'Loudness__dB', 'Speechiness']]
cleaned_top50.sort_values('Speechiness', ascending=False).head()
```

```python
# calculate the average speechiness
cleaned_top50['Speechiness'].mean()
```

```python
# calculate the average loudness
cleaned_top50['Loudness__dB'].mean()
```
