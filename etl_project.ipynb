```python
# Set up dependencies
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
# Reference the file where the CSV is located
data_df = 'data_files/data.csv'
# Import the data into a Pandas DataFrame
data_df = pd.read_csv(data_df)
data_df.head()
```

```python
# update df to keep specific columns and sort by speechiness
cleaned_data = data_df[['artists', 'name', 'loudness', 'speechiness', 'year']]
year2019_data = cleaned_data.loc[cleaned_data['year'] == 2019]
year2019_data.sort_values('speechiness', ascending=False).head()
```

```python
# calculate the average speechiness
year2019_data['speechiness'].mean()
```

```python
# calculate the average loudness
year2019_data['loudness'].mean()
```

```python
# Reference the file where the CSV is located
global_path = 'data_files/spotify_global_2019.csv'
# Import the data into a Pandas DataFrame
global_df = pd.read_csv(global_path)
global_df.head()
```

```python
# find all column names
global_df.columns
```

```python
# update df to keep specific columns and sort by speechiness
cleaned_global = global_df[['Rank', 'Track Name', 'Artist', 'loudness', 'speechiness']]
cleaned_global.set_index('Rank').sort_values('speechiness', ascending=False)
```

```python
# calculate the average speechiness
cleaned_global['speechiness'].mean().round(5)
```

```python
# calculate the average speechiness
cleaned_global['loudness'].mean().round(5)
```
