```python
#Dependencies and Setup
import pandas as pd
from sqlalchemy import create_engine
pd.options.display.float_format = '{:,.2f}%'.format
```

```python
#import CSV files
data_file = "data.csv"
Data_df = pd.read_csv(data_file)
Data_df.head()
```

```python
CleanData_df = Data_df [['year', 'artists', 'name', 'energy', 'danceability',  'speechiness']]
CleanData_df.head()
```

```python
# totalmusic_df = CleanData_df[(Data_df['year'] >= 1921) & (Data_df ['year'] <= 1949)]
# totalmusic_df = pd.DataFrame(totalmusic_df).groupby('year').mean()

# music_df = pd.concat(totalmusic_df).reset_index()

Decade20_df = CleanData_df[(Data_df['year'] >= 1920) & (Data_df ['year'] <= 1929)]
Decade20_df = pd.DataFrame(Decade20_df).groupby('year').mean()

Decade30_df = CleanData_df[(Data_df['year'] >= 1930) & (Data_df ['year'] <= 1939)]
Decade30_df = pd.DataFrame(Decade30_df).groupby('year').mean()

Decade40_df = CleanData_df[(Data_df['year'] >= 1940) & (Data_df ['year'] <= 1949)]
Decade40_df = pd.DataFrame(Decade40_df).groupby('year').mean()

music_decades = [Decade20_df, Decade30_df, Decade40_df]

total_music = pd.concat(music_decades).reset_index()
total_music
```

```python
total_music.to_csv("data_files/totalmusicavg.csv", index=False)
```
