```python
#Dependencies and Setup
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import CSV files
data_file = "data.csv"
Global_file = "Global2019.csv"
TopSongs_df = pd.read_csv(data_file)
GlobalSongs_df = pd.read_csv(Global_file)
```

```python
CleanGlobal_df = GlobalSongs_df [['Artist', 'Track Name', 'danceability', 'Artist_popularity']].copy()
FilteredGlobal_df = CleanGlobal_df[CleanGlobal_df['Artist_popularity'] <= 95]
FinalGlobal_df = FilteredGlobal_df.sort_values(by='Artist_popularity', ascending=False)
FinalGlobal_df.head(50)
```

```python
CleanData_df = TopSongs_df [['artists', 'danceability', 'popularity', 'name', 'year']].copy()
FilteredData_df = CleanData_df[CleanData_df['year'] == 2019]
FinalData_df = FilteredData_df.sort_values(by='popularity', ascending=False)
FinalData_df.head(50)
```

```python
password = "#Love3232"
postgres_conn_string = f"postgres:{password}" 
engine = create_engine(f"postgresql://{postgres_conn_string}@localhost:5432/ETL_DB")
```

```python
engine.table_names()
```

```python
FinalGlobal_df.to_sql(name='global_2019', con=engine, if_exists='replace', index=False)
```

```python
FinalData_df.to_sql(name='song_data', con=engine, if_exists='replace', index=False)
```
